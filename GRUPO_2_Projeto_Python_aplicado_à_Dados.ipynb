```python
spotify_tratado = spotify.rename(columns={
    "track_name": "nome_da_musica",
    "artist(s)_name": "nome_do_artista",
    "artist_count": "qtd_artistas",
    "released_year": "ano_lancamento",
    "released_month": "mes_lancamento",
    "released_day": "dia_lancamento",
    "in_spotify_playlists": "em_playlists_spotify",
    "in_spotify_charts": "nos_charts_spotify",
    "in_apple_playlists": "em_playlists_apple",
    "in_apple_charts": "nos_charts_apple",
    "in_deezer_playlists": "em_playlists_deezer",
    "in_deezer_charts": "nos_charts_deezer",
    "bpm": "batimentos_por_minuto",
    "nrgy": "energia",
    "dnce": "dancabilidade",
    "valence_%": "positividade",
    "acousticness_%": "acusticidade",
    "instrumentalness_%": "instrumentalidade",
    "liveness_%": "vivacidade",
    "speechiness_%": "fala"
}, inplace=False)

spotify_tratado.info()
```

```python
spotify_tratado.head()
```

```python
#true_remove = ["BPM110KeyAModeMajorDanceability53Valence75Energy69Acousticness7Instrumentalness0Liveness17Speechiness3"]
spotify_tratado = spotify_tratado.query(
 "streams != 'BPM110KeyAModeMajorDanceability53Valence75Energy69Acousticness7Instrumentalness0Liveness17Speechiness3'"
).assign(streams=spotify_tratado.streams.astype(int))
```

```python
#true_remove = ["BPM110KeyAModeMajorDanceability53Valence75Energy69Acousticness7Instrumentalness0Liveness17Speechiness3"]
spotify_tratado = spotify_tratado.query(
 "~ streams.str.contains('B')"
)#
spotify_tratado=spotify_tratado.assign(streams=spotify_tratado.streams.astype(int))

spotify_tratado.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")


colunas_numericas = spotify_tratado.select_dtypes(include=['int64', 'float64']).columns


plt.figure(figsize=(20, 20))
for i, coluna in enumerate(colunas_numericas, 1):
    plt.subplot(6, 3, i)
    sns.histplot(data=spotify_tratado, x=coluna, kde=True, color='skyblue')
    plt.title(f'DistribuiÃ§Ã£o de {coluna}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 20))
for i, coluna in enumerate(colunas_numericas, 1):
    plt.subplot(6, 3, i)
    sns.boxplot(data=spotify_tratado[coluna], color='orange')
    plt.title(f'Boxplot de {coluna}')
plt.tight_layout()
plt.show()
```

### Pergunta 1: *QUAL ARTISTA COM MAIOR QUANTIDADE DE MÃSICA TOCADAS?*

```python
spotify_tratado.groupby('nome_do_artista')['streams'].sum().nlargest(5).reset_index()
```

```python
top5_artistas = spotify_tratado.groupby('nome_do_artista')['streams'].sum().nlargest(5)

top5_artistas.plot(kind='barh', title='Top 5 Artistas com Mais Streams')
plt.xlabel('Total de Streams')
plt.ylabel('Artista')
plt.tight_layout()
plt.show()
```
