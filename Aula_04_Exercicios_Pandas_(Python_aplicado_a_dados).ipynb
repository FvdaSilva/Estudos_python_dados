Resumo

Leitura e Escrita de Dados:

- `pd.read_csv()`: Para ler dados de um arquivo CSV em um DataFrame.
- `pd.read_excel()`: Para ler dados de uma planilha Excel em um DataFrame.
- `pd.read_sql()`: Para ler dados de um banco de dados SQL em um DataFrame.
- `df.to_csv()`: Para salvar um DataFrame em um arquivo CSV.
- `df.to_excel()`: Para salvar um DataFrame em um arquivo Excel.

Estruturao e Inspeo de Dados:

- `pd.DataFrame()`: Cria um DataFrame a partir de dados em Python (listas, dicionrios, arrays, etc.).

- `df.head()`, `df.tail()`: Exibe as primeiras ou ltimas linhas do DataFrame.

- `df.shape`: Retorna a dimenso (nmero de linhas e colunas) do DataFrame.

- `df.info()`: Retorna informaes sobre o DataFrame, incluindo tipos de dados e valores no nulos.

- `df.describe()`: Gera estatsticas descritivas para as colunas numricas do DataFrame.

Seleo e Indexao:

- `df[column_name]`: Seleciona uma coluna do DataFrame.

Filtragem e Consultas:

- `df[df['column'] > value]`: Filtra o DataFrame com base em uma condio.
- `df.query('condition')`: Filtra o DataFrame usando uma string de consulta.
- `df.isnull(), df.notnull()`: Verifica quais valores so nulos ou no nulos.
- `df[df['column'].isin([value1, value2])]`: Filtra o DataFrame com base em valores especficos.

Operaes de Agregao:

- `df.groupby()`: Agrupa o DataFrame com base em uma ou mais colunas.
- `grouped_df.agg()`: Executa operaes de agregao em grupos de dados.
- `df.sum(), df.mean(), df.min(), df.max()`: Calcula a soma, mdia, valor mnimo e valor mximo das colunas numricas.

Operaes de Limpeza:

- `df.dropna()`: Remove linhas com valores nulos do DataFrame.
- `df.fillna(value)`: Preenche valores nulos com um valor especfico.
- `df.drop()`: Remove colunas ou linhas do DataFrame.
